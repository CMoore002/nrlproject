```python
import requests
import re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import random
import numpy as np
competitionType = '111'
seasonYear = 2010
roundNumber = 1
currentYear = 2023
```

```python
##Getting data for analysis
##Creating pandas dataframe for all data
allData = pd.DataFrame()

while(seasonYear < currentYear):
    ##Requesting page source
    page = requests.get('https://www.nrl.com/draw/?competition={}&round={}&season={}'.format(competitionType, roundNumber, seasonYear))

    ##Parsing with BeautifulSoup
    soup = BeautifulSoup(page.content, 'html.parser')
    
    ##Getting data from page. Finding the class and then searching q-data in the div element
    roundData = str(soup.find(class_="u-spacing-mt-24")['q-data'])
    
    allData = pd.concat([allData, dataSplit(roundData).T])
    roundNumber += 1
    if allData[0].iloc[len(allData[0]) - 1] == 'GrandFinal':
        seasonYear += 1
        roundNumber = 1
        
##Renaming column headers
allData = allData.rename(columns = {0:'Round', 1:'Date', 2:'Home Team', 3:'Home Score', 4:'Away Team', 5:'Away Score', 6:'Season'})

##Creating margin column to show the score margin with home team as point of reference
allData['Margin'] = allData['Home Score'].astype(int) - allData['Away Score'].astype(int)

##Resetting the row indices and then dropping the old index from dataframe
allData = allData.reset_index().drop('index', axis=1)

```

```python
##Defining functions
def removeCharacters(string):
    """
    Function returns the string without special characters
    """
    return re.sub('\W+', '', string)

def dataSplit(roundData):
    """
    This is a function that returns Round Number, Home Team, Home Team Score, Away Team, Away Team Score
    and Date of Match given the data(for the specified year?)
    """
    ##Splitting the data into a list for each match. matchCentreUrl is the split to divide each
    ##match into it's own list index
    matchList = []
    for text in roundData.split('matchCentreUrl'):
        matchList.append(text)
    
    ##Finding Round Number. Finds first occurrence of '"roundTitle":' and splits. Split again to get 
    ##round number
    roundList = []
    for x in range(1, len(matchList)):
        roundList.append(removeCharacters(roundData.split('"roundTitle":')[1].split(',')[0]))
    
    ##Finding Date of Match
    dateData = []
    seasonData = []
    for x in range(1, len(matchList)):
        dateData.append(matchList[x].split('"kickOffTimeLong":')[1].split(',')[0].split('T')[0].replace('"', ''))
        seasonData.append(matchList[x].split('"kickOffTimeLong":')[1].split(',')[0].split('T')[0].replace('"', '').split('-')[0])
    ##Finding Team Name Data
    ##the index for teamData is (x-1) because the loop starts at 1
    teamData = []
    homeList = []
    awayList = []
    for x in range(1, len(matchList)):
        teamData.append([removeCharacters(y.split(':')[1]) for y in matchList[x].split(',') if "nickName" in y])
        homeList.append(teamData[x-1][0])
        awayList.append(teamData[x-1][1])
    
    ##Finding score Data
    ##the index for scoreData is (x-1) because the loop starts at 1
    scoreData = []
    homeScore = []
    awayScore = []
    for x in range(1, len(matchList)):
        scoreData.append([removeCharacters(y.split(':')[1]) for y in matchList[x].split(',') if "score" in y])
        homeScore.append(scoreData[x-1][0])
        awayScore.append(scoreData[x-1][1])        
    return pd.DataFrame([roundList, dateData, homeList, homeScore, awayList, awayScore, seasonData])

##Creating function for updating the ladder
def dictionaryUpdate(year):
    """
    This function updates the dictionary ladderData with FOR, AGAINST, POINTS for each team
    for the year entered. Data is gathered from allData pandas dataframe.
    """

    ##192 is just until the end of the season before the finals start for 2010.
    for x in range(192):
        ##Update Home Team
        homeTeam = allData.iloc[x]['Home Team']
        
        ##0 index is for FOR
        ladderData[homeTeam][0] += int(allData.iloc[x]['Home Score'])
        
        ##1 index is for AGAINST
        ladderData[homeTeam][1] += int(allData.iloc[x]['Away Score'])
        
        ##2 index is for POINTS
        if int(allData.iloc[x]['Margin']) > 0:
            ladderData[homeTeam][2] += 2
        elif int(allData.iloc[x]['Margin'] == 0):
            ladderData[homeTeam][2] += 1

            
        ##Update Away Team
        awayTeam = allData.iloc[x]['Away Team']
        
        ##0 index is for FOR
        ladderData[awayTeam][0] += int(allData.iloc[x]['Away Score'])
        
        ##1 index is for AGAINST
        ladderData[awayTeam][1] += int(allData.iloc[x]['Home Score'])
        
        ##2 index is for POINTS
        if int(allData.iloc[x]['Margin']) < 0:
            ladderData[awayTeam][2] += 2
        elif int(allData.iloc[x]['Margin'] == 0):
            ladderData[awayTeam][2] += 1  
            
    return
```
